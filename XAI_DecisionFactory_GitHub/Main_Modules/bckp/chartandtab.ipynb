```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Load data ---
df = pd.read_csv("XDF_analytical_output/PROFILE_STATS.csv")

# --- Config ---
colors = ["#1a5c2a" if str(p) == "5" else "#9e9e9e" for p in df["Profile_ID"]]

# --- Plot ---
fig, ax = plt.subplots(figsize=(8, 5))

bars = ax.bar(
    df["Profile_ID"].astype(str),
    df["HP"],
    color=colors,
    edgecolor="white",
    linewidth=0.8
)

# --- Labels on bars ---
for bar, val in zip(bars, df["HP"]):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 0.005,
        f"{val:.2f}",
        ha="center",
        va="bottom",
        fontsize=10,
        color="#333333"
    )

# --- Styling ---
ax.set_title("Propensity by Profile", fontsize=13, fontweight="bold", pad=15)
ax.set_xlabel("Profile ID", fontsize=11)
ax.set_ylabel("HP Rate", fontsize=11)
ax.set_ylim(0, df["HP"].max() * 1.2)
ax.spines[["top", "right"]].set_visible(False)
ax.yaxis.grid(True, linestyle="--", alpha=0.5)
ax.set_axisbelow(True)

# --- Legend ---
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor="#1a5c2a", label="Profile 11"),
    Patch(facecolor="#9e9e9e", label="Other Profiles")
]
ax.legend(handles=legend_elements, frameon=False)

plt.tight_layout()
plt.savefig("hp_rate_by_profile.png", dpi=150)
plt.show()

import os

output_dir = "XDF_analytical_output"
os.makedirs(output_dir, exist_ok=True)

plt.tight_layout()
plt.savefig(os.path.join(output_dir, "hp_rate_by_profile.png"), dpi=150, bbox_inches="tight")
plt.show()
```
