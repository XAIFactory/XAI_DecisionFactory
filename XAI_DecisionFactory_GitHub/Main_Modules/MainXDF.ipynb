```python
from ProfileGen import ds_profile_generator
from ContextAgent import execute_agent_context
from ProfilerAgent import run_profile_agent
```

```python
def main():

    # ---------------------------------------------------------
    # STEP 1: Data Science profiling module
    # Generates customer profiles, SHAP importance,
    # clustering, decision rules, profile summary stats,
    #and odds ratios
    # ---------------------------------------------------------
    res1 = ds_profile_generator(
            work_directory="/XAI_DecisionFactory_GitHub",  # Root project directory
            file_path="XDF_abt_input/ActionForge.csv",  # Input dataset (CSV format)
            sep=";",                                     # CSV column separator
            decimal=",",                                 # Decimal separator
            id_col="IDCLI",                              # Unique identifier column
            target_score_col="score",                    # Model score column
            threshold=10,                                # Top 10% = High Propensity
            min_profile_size=0.06,                       # Minimum profile size (% of total population)
            eps=0.25,                                     # DBSCAN neighborhood radius
            output_path="XDF_analytical_output"          # Output folder
)

    # ---------------------------------------------------------
    # STEP 2: Context ingestion agent
    # Loads business, general and feature-level context
    # to enrich the interpretation layer
    # ---------------------------------------------------------
    res2 = execute_agent_context(
        work_directory="/XAI_DecisionFactory_GitHub",     #work directory where the folder containing the "main_modules" folder is stored
        general_csv="XDF_context_infos/general_context.csv",   # General contextual information
        business_csv="XDF_context_infos/biz_context.csv",      # Business context and domain knowledge
        features_csv="XDF_context_infos/features_context.csv",  # Feature descriptions and human-readable meanings
        ollama_url = "http://localhost:11434/api/chat", #it's standard host of local ollama (change just if required)
        llm = "phi3:mini" #ollama local installed model (eg if you have mintsral and GPU you can use it)
    )

    # ---------------------------------------------------------
    # STEP 3: Profile interpretation agent
    # Combines analytical outputs with context to generate
    # explainable and actionable insights at profile level
    # ---------------------------------------------------------
    res3 = run_profile_agent(
        work_directory       = "/XAI_DecisionFactory_GitHub",
        oddratio_csv         = "XDF_analytical_output/oddsratio.csv",
        profile_features_csv = "XDF_analytical_output/profile_features.csv",
        profile_stats_csv    = "XDF_analytical_output/profile_stats.csv",
        profile_rules_csv    = "XDF_analytical_output/profile_rules.csv",
        shap_importance_json = "XDF_analytical_output/shap_importance.json",
        context_memory       = "context_memory.json",
        sep                  = ",",
        llm                  = "phi3:mini",
        ollama_url           = "http://localhost:11434/api/chat"
    )

    # ---------------------------------------------------------
    # OUTPUT RESULTS
    # ---------------------------------------------------------
    print("Step 1 - Profiles Generation:")
    print(res1)

    print("\nStep 2 - Context Agent Output:")
    print(res2)

    print("\nStep 3 - Profiler Agent Output:")
    print(res3)


if __name__ == "__main__":
    main()
```
