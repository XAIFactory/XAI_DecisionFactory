```python
from ProfileGen import ds_profile_generator
from ContextAgent import execute_agent_context
from ProfilerAgent import run_profile_agent
```

```python
def main():

    # ---------------------------------------------------------
    # STEP 1: Data Science profiling module
    # Generates customer profiles, SHAP importance,
    # clustering, decision rules, profile summary stats,
    #and odds ratios
    # ---------------------------------------------------------
    res1 = ds_profile_generator(
        work_directory="XAI_DecisionFactory", #work directory where the folder containing the "main" is stored
        file_path="XDF_abt_input/ActionForge.csv",        # Input dataset (CSV format)
        sep=";",                                         # CSV column separator
        decimal=",",                                     # Decimal separator
        id_col="IDCLI",                                  # Unique identifier column
        target_score_col="score",                        # Model score column
        threshold=10,                                    # Percentile threshold for HP/LP definition (10: top10% will be marked as HP(high Propensity)"
        min_profile_size=0.06,                           # Minimum profile size (as % of total population)
        eps=0.25,                                        # DBSCAN neighborhood radius - clustering parameters
        output_path="XDF_analytical_output"               # Output directory for all generated artifacts
    )

    # ---------------------------------------------------------
    # STEP 2: Context ingestion agent
    # Loads business, general and feature-level context
    # to enrich the interpretation layer
    # ---------------------------------------------------------
    res2 = execute_agent_context(
        work_directory="XAI_DecisionFactory",     #work directory where the folder containing the "main" is stored
        general_csv="XDF_context_infos/general_context.csv",   # General contextual information
        business_csv="XDF_context_infos/biz_context.csv",      # Business context and domain knowledge
        features_csv="XDF_context_infos/features_context.csv"  # Feature descriptions and human-readable meanings
    )

    # ---------------------------------------------------------
    # STEP 3: Profile interpretation agent
    # Combines analytical outputs with context to generate
    # explainable and actionable insights at profile level
    # ---------------------------------------------------------
    res3 = run_profile_agent(
        work_directory="XAI_DecisionFactory", #work directory where the folder containing the "main" is stored
        oddratio_csv="XDF_analytical_output/oddsratio.csv",          # Odds ratios from logistic model
        profile_features_csv="XDF_analytical_output/PROFILE_FEATURES.csv",  # Decision tree feature importance
        profile_stats_csv="XDF_analytical_output/PROFILE_STATS.csv",        # Aggregated statistics per profile
        profile_rules_csv="XDF_analytical_output/PROFILE_RULES.csv",        # Decision tree rules per profile
        shap_importance_json="XDF_analytical_output/shap_importance.json",  # Global SHAP importance
        context_memory="context_memory.json",                             # Context generated by previous agent
        sep=","                                                           # CSV separator
    )

    # ---------------------------------------------------------
    # OUTPUT RESULTS
    # ---------------------------------------------------------
    print("Step 1 - Profiler Output:")
    print(res1)

    print("\nStep 2 - Context Agent Output:")
    print(res2)

    print("\nStep 3 - Profile Agent Output:")
    print(res3)


if __name__ == "__main__":
    main()
```
